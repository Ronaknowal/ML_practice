```python
def gradient_descent(x,y):
    m_curr=b_curr=0
    iterations=100
    # rather than no of iteration we can also end the for loop if the cost is below a certain amount.
        learning_rate=0.0002
        n=len(x)
    #     plt.scatter(x,y,color='red',marker='+',linewidth='5')
    for i in range(iterations):
        y_pred=m_curr*x+b_curr
#         plt.plot(x,y_pred,color='green')
        cost=(1/n)*sum([val**2 for val in (y-y_pred)])
        m_der=-(2/n)*sum(x*(y-y_pred))
        b_der=-(2/n)*sum(y-y_pred)
        m_curr=m_curr-learning_rate*m_der
        b_curr=b_curr-learning_rate*b_der
        print("m {}, b {}, cost {}, iteration {}".format(m_curr,b_curr,cost,i))
```

```python
gradient_descent(df.math,df.cs)
```
